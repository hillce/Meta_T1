```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("./TrainingLogs/MU_INVT1_SD_INVTIMES/Training_Loss.json","r") as f:
    valLossInv = json.load(f)

valLossInvFlat = [x for sublist in valLossInv for x in sublist]
valLossInvMean = [np.mean(sublist) for sublist in valLossInv]

with open("./TrainingLogs/MU_INVT1_SD_0INVTIMES/Training_Loss.json","r") as f:
    valLoss0Inv = json.load(f)

valLoss0InvFlat = [x for sublist in valLoss0Inv for x in sublist]
valLoss0InvMean = [np.mean(sublist) for sublist in valLoss0Inv]

with open("./TrainingLogs/MU_INVT1_SD_INVTIMES_BMI/Training_Loss.json","r") as f:
    valLossInvBMI = json.load(f)

valLossInvBMIFlat = [x for sublist in valLossInvBMI for x in sublist]
valLossInvBMIMean = [np.mean(sublist) for sublist in valLossInvBMI]

with open("./TrainingLogs/MU_INVT1_SD_0INVTIMES_BMI/Training_Loss.json","r") as f:
    valLoss0InvBMI = json.load(f)

valLoss0InvBMIFlat = [x for sublist in valLoss0InvBMI for x in sublist]
valLoss0InvBMIMean = [np.mean(sublist) for sublist in valLoss0InvBMI]

plt.plot(valLossInvFlat,"b",alpha=0.5)
plt.plot(valLoss0InvFlat,"r",alpha=0.5)
plt.plot(valLossInvBMIFlat,"g",alpha=0.5)
plt.plot(valLoss0InvBMIFlat,"pink",alpha=0.5)
plt.xlabel("Step")
plt.ylabel("Loss")
plt.show()

plt.plot(valLossInvMean,"b",alpha=0.5,label="Inversion Times")
plt.plot(valLoss0InvMean,"r",alpha=0.5,label="Zeroed Inversion Times")
plt.plot(valLossInvBMIMean,"g",alpha=0.5,label="Inversion Times, BMI Minimisation")
plt.plot(valLoss0InvBMIMean,"pink",alpha=0.5,label="Zeroed Inversion Times, BMI Minimisation")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
with open("./TrainingLogs/MU_INVT1_SD_INVTIMES/Val_Loss.json","r") as f:
    valLossInv = json.load(f)

with open("./TrainingLogs/MU_INVT1_SD_0INVTIMES/Val_Loss.json","r") as f:
    valLoss0Inv = json.load(f)

with open("./TrainingLogs/MU_INVT1_SD_INVTIMES_BMI/Val_Loss.json","r") as f:
    valLossInvBMI = json.load(f)

with open("./TrainingLogs/MU_INVT1_SD_0INVTIMES_BMI/Val_Loss.json","r") as f:
    valLoss0InvBMI = json.load(f)

valLossInv = np.array(valLossInv)
valLoss0Inv = np.array(valLoss0Inv)
valLossInvBMI = np.array(valLossInvBMI)
valLoss0InvBMI = np.array(valLoss0InvBMI)

valLossInvFlat = valLossInv.flatten()
valLoss0InvFlat = valLoss0Inv.flatten()
valLossInvBMIFlat = valLossInvBMI.flatten()
valLoss0InvBMIFlat = valLoss0InvBMI.flatten()

valLoss0InvMean = np.mean(valLoss0Inv,axis=1)
valLossInvMean = np.mean(valLossInv,axis=1)
valLossInvBMIMean = np.mean(valLossInvBMI,axis=1)
valLoss0InvBMIMean = np.mean(valLoss0InvBMI,axis=1)

plt.plot(valLossInvFlat,"b",alpha=0.5)
plt.plot(valLoss0InvFlat,"r",alpha=0.5)
plt.plot(valLossInvBMIFlat,"g",alpha=0.5)
plt.plot(valLoss0InvBMIFlat,"pink",alpha=0.5)
plt.xlabel("Step")
plt.ylabel("Loss")
plt.show()

plt.plot(valLossInvMean,"b",alpha=0.5,label="Inversion Times")
plt.plot(valLoss0InvMean,"r",alpha=0.5,label="Zeroed Inversion Times")
plt.plot(valLossInvBMIMean,"g",alpha=0.5,label="Inversion Times, BMI Minimisation")
plt.plot(valLoss0InvBMIMean,"pink",alpha=0.5,label="Zeroed Inversion Times, BMI Minimisation")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
